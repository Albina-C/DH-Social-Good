```python
# import packages here
import pandas as pd
import numpy as np
import sqlite3 as sq
from scipy import stats
import matplotlib.pyplot as plt
```

```python
crime=pd.read_csv('Mar5Crime.csv')
```

```python
crime.head()
```

```python
crime= crime.drop(columns=['Unnamed: 0'])
```

```python
crime.head()
```

```python
crime.describe()
```

Creating dataframe for crimes WITHOUT location data

```python
df1= crime[crime['Missing_Data']==1] #dataframe that only consists of missing locations
```

```python
df1.head()
df1.reset_index()
```

```python
ser1 = df1.hour.value_counts()
#print(ser1)
list1 = list(ser1)
#print(list1) # count values of each hours (not sorted)
```

```python
hours = list(ser1.index) # hours of the value_counts list
```

```python
d={'hour':hours, 'count':list1}
df = pd.DataFrame(data=d) #create a dataframe with hours and count
```

```python
frequency_missing = df.sort_values('hour')
frequency_missing = frequency_missing.reset_index()
frequency_missing = frequency_missing.drop(columns=['index'])
```

Creating dataframe for crimes WITH location data

```python
df2= crime[crime['Missing_Data']==0]
df2.head()
df2.reset_index()
```

```python
ser2 = df2.hour.value_counts()
list2 = list(ser2) # count values of each hours (not sorted)
```

```python
hours2 = list(ser2.index)
```

```python
d2={'hour':hours2, 'count':list2}
df2 = pd.DataFrame(data=d2) #create a dataframe with hours and count
```

```python
frequency_normal = df2.sort_values('hour')
frequency_normal = frequency_normal.reset_index()
frequency_normal = frequency_normal.drop(columns=['index'])
```

```python
frequency_normal
```

```python
merged_frequency= frequency_missing.merge(frequency_normal,how='inner',on='hour')
```
