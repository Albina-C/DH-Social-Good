These are the two data sets we were provided with from the Boston files. I have loaded them all for you guys using the packages that we learned (pandas). We are now ready to start cleaning!

```python
# here is our loaded files
crime= pd.read_csv('crime - first six months.csv', engine='python')
```

```python
new_colnames = [x.lower().replace(' ', '_') for x in crime.columns]
crime.columns= new_colnames
crime.head()
```

```python
crime['shooting'] = crime['shooting'].fillna(False)
```

```python
crime.head()
```

```python
crime['district'] = crime['district'].fillna('N/A')
```

```python
crime['reporting_area'] = crime['reporting_area'].replace(' ','N/A')
```

```python
crime['ucr_part'] = crime['ucr_part'].fillna('N/A')
```

```python
crime['street'] = crime['street'].fillna('N/A')
```

```python
crime['lat'] = crime['lat'].fillna('N/A')
crime['long'] = crime['long'].fillna('N/A')
```
