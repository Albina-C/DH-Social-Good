These are the two data sets we were provided with from the Boston files. I have loaded them all for you guys using the packages that we learned (pandas). We are now ready to start cleaning!

```python
# here is our loaded files
crime= pd.read_csv('crime_last_six.csv', engine='python')
```

```python
new_colnames = [x.lower().replace(' ', '_') for x in crime.columns]
crime.columns= new_colnames
crime.head()
```

We are dropping columns that are not needed. 

```python
crime=crime.drop(columns=['offense_code','district','reporting_area','ucr_part','year'], axis=1)
crime.head(100)
```

Replacing NaN values with 0 (no shooting) and 'Y' values with 1 (shooting)

```python
crime["shooting"]=crime["shooting"].fillna(0)
crime["shooting"] = crime["shooting"].replace(to_replace="Y",value=1)
crime.head(1287)
```

We decided to not use the shooting column anymore, so we drop it in the code below. 

```python
crime=crime.drop(columns=['shooting'], axis=1)
crime.head(500)
```

```python
crime['Missing_Data']=[0]*len(crime['lat'])
```

```python
crime.loc[crime.location == '(-1.00000000, -1.00000000)', "Missing_Data"] = 1
crime.loc[crime.location == '(0.00000000, 0.00000000)', "Missing_Data"] = 1
```
